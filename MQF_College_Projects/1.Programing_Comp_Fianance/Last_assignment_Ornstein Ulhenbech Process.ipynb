```python
import math
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import Image

plt.style.use('seaborn')
sns.set_palette('cubehelix')
plt.rcParams['figure.figsize'] = [8, 4.5]
plt.rcParams['figure.dpi'] = 300
```

```python
def OU_process(t_0,t_end, length,theta,mu,sigma):
    t = np.linspace(t_0,t_end,length) 
    dt = np.mean(np.diff(t))
    y = np.zeros(length)
    y0 = np.random.normal(loc=0.0,scale=1.0)
    drift = lambda y,t: theta*(mu-y) # defining drift term
    diffusion = lambda y,t: sigma # defining diffusion term
    noise = np.random.normal(loc=0.0,scale=1.0,size=length)*np.sqrt(dt) #defining noise process
    
    for i in range(1,length): 
        y[i] = y[i-1] + drift(y[i-1],i*dt)*dt + diffusion(y[i-1],i*dt)*noise[i]
   
    return y  
```

```python
np.random.seed(1)
plt.plot(t, OU_process(t_0 = 1, t_end = 5, length = 1000, theta = 2.0, mu = 1.0, sigma = 0.3))
plt.plot(t, OU_process(t_0 = 1, t_end = 5, length = 1000, theta = 2.0, mu = 3.0, sigma = 0.3))
plt.plot(t, OU_process(t_0 = 1, t_end = 5, length = 1000, theta = 2.0, mu = 5.0, sigma = 0.3))
plt.legend(["Mu = 1.0", "Mu = 3.0", "Mu = 5.0"], loc = "lower right")
```
