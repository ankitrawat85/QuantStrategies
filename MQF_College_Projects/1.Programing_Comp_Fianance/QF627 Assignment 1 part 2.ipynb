### Extract some data from my cluster

> Extract data from the cluster

```python
def extract_data(database, collection):
    
    # Access to the database
    db = client[database]
    
    # Look into the collection you want check at that database
    coll = db[collection]
    
    # Ignore the '_id' and 'plot' item
    data = coll.find({}, {'_id' : 0, 'plot' : 0})
    
    return data
```

> Iterate the data to a DataFrame

```python
import pandas as pd
```

```python
def mongo_to_df(database, collection):
    my_df = pd.DataFrame(list(extract_data(database, collection)))
    df = my_df.set_index('year')
    return df
```
