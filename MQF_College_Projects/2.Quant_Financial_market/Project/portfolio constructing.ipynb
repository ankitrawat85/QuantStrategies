```python
data = df_return.iloc[1:, ]
data
```

```python
R_array = np.array([data.mean()]).T
R_array
```

```python
cov = data.cov()
```

```python
cov_inv = np.linalg.inv(cov)
```

```python
e = np.ones((10, 1), dtype = int)
```

```python
Alpha = R_array.T @ cov_inv @ e
Zeta = np.dot(np.dot(R_array.T, cov_inv), R_array)
Delta = np.dot(np.dot(e.T, cov_inv), e)
```

```python
# riskfree rate
Rf = 0.0013
Rf
```

```python
Rmv = Alpha/Delta
```

```python
# Get the return and variance of the tangency portfolio
Rtg = (Alpha * Rf - Zeta)/(Delta * Rf - Alpha)
Var_tg = (Zeta - 2*Alpha*Rf + Delta*(Rf ** 2))/((Delta ** 2) * ((Rf - Rmv)**2))
print(Rtg)
print(Var_tg)
```

```python
# Sharpe ratio of the tangency portfolio
sharpe_ratio = np.sqrt(Zeta - 2 * Alpha * Rf + Delta * (Rf ** 2))
print('The Sharpe ration of the tangency portfolio is ' + str(round(sharpe_ratio[0][0], 4)))
```

```python
lambda_ = (Rtg - Rf)/(Zeta - 2 * Alpha * Rf + Delta * (Rf ** 2))
weight = lambda_ * cov_inv @ (R_array - Rf * e)
weight
```

```python
# Expected return deviation
R = np.arange(0, 0.1, 0.001)
```

```python
# Variance and Std of return for portfolio on minimum-tracking-error frontier
Var = 1/Delta + (R - Alpha/Delta)**2 * Delta / (Zeta * Delta - Alpha ** 2)
std = np.sqrt(Var)
std = std[0]
```

```python
# Variance of return for portfolio with riskless asset when expected return is 0.1%
var_portfolio = (0.1 - Rf)**2 /(Zeta - 2 * Alpha * Rf + Delta * (Rf ** 2))
std_portfolio = np.sqrt(var_portfolio)
```

```python
plt.figure(figsize = (15, 8))
plt.plot(std, R)
plt.plot([0, std_portfolio],[Rf, 0.1], c = 'red')
plt.xlabel('Std of return', fontsize=20)
plt.ylabel('Expected Return', fontsize=20)
plt.title('Efficient Frontier', fontsize=40)
#plt.legend(frameon = False, loc = 'lower center', ncol = 2, fontsize = 16);
#plt.xlim(0,0.25)
plt.ylim(0,0.1);
```
